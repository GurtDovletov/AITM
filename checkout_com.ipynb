# Dropping with most categorical columns and most missing value columns

```python
X_full.drop(["network_category"], axis = 1, inplace = True)
X_full.drop(["proxy_ip"], axis = 1, inplace = True)
X_full.drop(["mouse_speed"], axis = 1, inplace = True)
X_full.drop(["cursor_velocity"], axis = 1, inplace = True)
X_full.drop(["scrolling_speed"], axis = 1, inplace = True)
X_full.drop(["battery_score"], axis = 1, inplace = True)
X_full.drop(["bot_score"], axis = 1, inplace = True)
X_full.drop(["network_dist"], axis = 1, inplace = True)
X_full.drop(["mouse_hor_ver_perc"], axis = 1, inplace = True)
X_full.drop(["phone_tilt"], axis = 1, inplace = True)
X_full.drop(["bh_score"], axis = 1, inplace = True)
X_full.drop(["country_code"], axis = 1, inplace = True)
X_full.drop(["transaction_date"], axis = 1, inplace = True) 
```

# Finding number of unique values each column

```python
s = (X_full.dtypes == 'object')
object_cols = list(s[s].index)

print(object_cols)


for col in object_cols:
    print(col+"____", X_full[col].nunique())
```

# Using one_hot_encoder for filling categorical values

```python
from sklearn.preprocessing import OneHotEncoder
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols_full = pd.DataFrame(OH_encoder.fit_transform(X_full[object_cols]))
OH_cols_full.index = X_full.index
num_X_full = X_full.drop(object_cols, axis=1)
OH_X_full = pd.concat([num_X_full, OH_cols_full], axis=1)
OH_X_full.columns = OH_X_full.columns.astype(str)
```

```python
OH_X_full.head()
```

# Number of missing values each column

```python
OH_X_full.isna().sum()
```

```python
cols_with_missing = [col for col in OH_X_full.columns if OH_X_full[col].isnull().any()]
print(cols_with_missing)
```

# Using extansion imputer method filling the missing values

```python
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer()

OH_X_full_plus = OH_X_full.copy()

for col in cols_with_missing:
    OH_X_full_plus[col + "with_missing"] = OH_X_full_plus[col].isnull()
    
imputed_OH_X_full_plus = pd.DataFrame(my_imputer.fit_transform(OH_X_full_plus))

imputed_OH_X_full_plus.columns = OH_X_full_plus.columns
```

```python
imputed_OH_X_full_plus.head()
```

# Seeing fraud is imbalance column
## Mojority "'0" and minority "1"

```python
counts = imputed_OH_X_full_plus['fraud'].value_counts()
print("0", counts[0])
print("1", counts[1])
```

# Randomly downsampling minority class inorder to making it balance

## I did this because if I undersampling my data would be very high

```python
# Separate minority and majority classes
minority_class = imputed_OH_X_full_plus[imputed_OH_X_full_plus['fraud'] == 1]
majority_class = imputed_OH_X_full_plus[imputed_OH_X_full_plus['fraud'] == 0]

# Downsample majority class to match the number of minority class
majority_downsampled = majority_class.sample(n=len(minority_class), random_state=42)

# Concatenate minority class and downsampled majority class
balanced_df = pd.concat([minority_class, majority_downsampled])

# Shuffle the balanced dataset
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

# Here indexing
## payment_id chaning it index because of increases one by one

```python
balanced_df.set_index('payment_id', inplace=True)
```

```python
balanced_df.head()
```

```python
balanced_df.shape
```

# Splitting data into train and valid datasets

```python
from sklearn.model_selection import train_test_split

y = balanced_df.fraud
X = balanced_df.drop(["fraud"], axis = 1)
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
from sklearn.metrics import mean_absolute_error
# from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

# my_model = RandomForestRegressor(random_state=0)
my_model = XGBRegressor()

my_model.fit(X_train, y_train)
my_model.fit(X_train, y_train)
y_pred = my_model.predict(X_valid)
```

# Accuracy score cannot calculate continous because we are dealing binary thats why we converting it into the continous

```python
from sklearn.metrics import accuracy_score

# Convert continuous predictions to binary
y_pred_binary = [1 if pred >= 0.5 else 0 for pred in y_pred]

# Calculate accuracy
accuracy = accuracy_score(y_valid, y_pred_binary)
print("Accuracy:", accuracy)
```
